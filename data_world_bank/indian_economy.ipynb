# Introduction to GDP

## What is Gross Domestic Product (GDP) ?

- A measure of value of all goods and services produced in a country
- Reflects economic activity
- Calculating GDP: 3 different Ways (Production, Income, Expenditure)

## Nominal vs. Real GDP
- Nominal GDP: GDP at current prices
- Real GDP: GDP at constant prices relative to a base year
- Example: Consider increase in Petrol prices in 2020 and 2021
- Real GDP = Nominal GDP - Inflation
- When govt. says 5 trillion dollar GDP one should ask whether nominal or real



## GDP in itself is not very representative of the economy

## Per Capita GDP
- Large population means large GDP
- Use Per Capita GDP =  ${\frac{\text{GDP}}{\text{Population}}}$

## Per Capita GDP Purchasing Power Parity (PPP) 
- Purchasing power parity: Accounts for difference in prices in different countries
- Example: 
  1. Buying 1 Kg rice in Sweden (25 SEK) vs. India (50 INR) 
  2. Renting Apartment in Sweden (7000 SEK) vs. India (14000 INR)
- A measure of effective exchange rate for an individual

## India's GDP growth rate
GDP growth rate = $\frac{\Delta G}{G}$

Mean growth rate from 1960-1990: 4.2%


Mean growth rate from 1991-2019: 6.4%

- Why large fluctuations in growth rate (sustainability is less)?

```python
df3=pd.read_csv("gdp_growth.csv",header=None,skiprows=1)
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(df3[0],df3[1],marker="o")
ax.set_xlabel("Year"); plt.ylabel("GDP growth rate in %")
ax.annotate("1991 reforms", xy=(1992,0.2),size=16,color='red')
std1=df3[1][1:30].std(); std2=df3[1][31:].std(); avg1=df3[1][1:30].mean();avg2=df3[1][31:].mean()
ax.vlines(ymin=-8,ymax=10,x=1991, color='red'); ax.set_title("India's GDP growth rate")
ax.annotate(f"sigma={round(std1,1)}%", xy=(1960,-7),size=12); 
ax.annotate(f"sigma={round(std2,1)}%", xy=(1992,-7),size=12)
ax.annotate(f"mean={round(avg1,1)}%", xy=(1960,-5),size=12); 
ax.annotate(f"mean={round(avg2,1)}%", xy=(1992,-5),size=12)
plt.savefig("Fig2.pdf",dpi=600)
plt.show()
```

## How can India can achieve high GDP growth rate ?


GDP growth rate $ = \frac{\Delta G}{G} = \frac{\Delta K}{G} * \frac {\Delta G}{\Delta K}$ 
= $ \text {investment-GDP ratio} * \eta $

2 factors:  
1. $\Delta K$ is increase in capital investment in a given year
2. $ \eta $ is efficiency of capital use

### How government policies influence $\Delta{K}$ ?
$${\Delta K}=\Delta{K_{private}} + \Delta{K_{public}} $$

- Private investment: Spending your money for your company (more responsibility)
- Public investment: Giving your money to govt. who gives it to someone else for spending (less reponsibility)
- Limitations of public investment:
1. High Public Investment Means Heavy Taxes on Poor and Middle Class
2. Inefficient 
3. Corruption and Favoritism
- Drawbacks of private investment: 
1. Labor Exploitation ( )
2. Lack of employee welfare

```python
invest=pd.read_csv("capital_investment_innovation.csv")
invest=invest.fillna(0)
fig, ax = plt.subplots(1,2,figsize=(8,4))
ax[0].plot(invest['Year'],invest['dKprivate'],marker="o")
ax[0].plot(invest['Year'],invest['dKpublic'],marker="o")
ax[0].set_xlabel("Year")
ax[0].set_ylabel(r'$\Delta{K}$ (as % of GDP)')
ax[0].set_title("India's $\Delta{K}/GDP$")
ax[0].vlines(ymin=5,ymax=35,x=1991, color='red')
ax[0].annotate("1991 reforms", xy=(1960,30),size=16,color='red')
ax[0].legend([r'$\Delta{K_{private}}$',r'$\Delta{K_{public}}$'],loc=(0.6,0.8),frameon=False)
ax[0].vlines(ymin=5,ymax=30,x=2012, color='green')
ax[0].annotate("2012", xy=(2012,28),size=16,color='red')

eodb=pd.DataFrame()
eodb['year']=[i for i in range(2011,2020)]
eodb['rank']=[132,131,134,134,131,130,100,77,63]
ax[1].set_ylim(30,150)
ax[1].set_xticks(eodb['year'])
ax[1].tick_params(axis='x', rotation=90)
ax[1].bar(eodb['year'],eodb['rank'])
ax[1].set_title("EODB Rank")
ax[1].set_xlabel("Year")
ax[1].set_ylabel('EODB Rank')

for i in range(len(eodb)):
    plt.annotate(eodb['rank'][i], xy=(eodb['year'][i],eodb['rank'][i]),ha='center', va='bottom')
plt.tight_layout()
plt.savefig("Fig3.pdf",dpi=600)
```

### How government policies influence ${\eta}$ ?

#### Exogeneous Economic Growth Model
- Developed by Prof. Robert Solow (1981 Nobel Memorial Prize in Economics)
- GDP growth rate is limited by Law of diminishing marginal returns
- Only 35 % growth is due to ${\Delta K}$; 65% growth rate is due to $\eta$ (which is a function of innovation)
- Innovation is essential to increase $\eta$ 
- But considered as an external factor of the economic model

```python
dim=pd.read_csv("diminish.csv",header=None)
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(dim[0],dim[1],marker="o")
ax.plot(dim[0],dim[2],marker="o")
ax.set_xlabel("Input (Labour & Capital)")
ax.set_ylabel("Output (Units)")
ax.annotate("Innovation", xy=(6.9,45), color='r',size=12)
ax.set_title("Diminishing Marginal Returns")
ax.vlines(ymin=40,ymax=56,x=6.8, color='red')
plt.savefig("Fig4.pdf",dpi=600)
plt.show()
```

### How government policies influence ${\eta}$ ?

#### Endogeneous Economic Growth Model
- Prof. Paul Romer (2018 Nobel Memorial Prize in Economics, former IMF president)
- Integrates innovation into exogeneous economic model

### $$\eta = \eta_{business} + \eta_{tech}$$

- Diminishing returns does not apply to human/knowledge capital
- Sustainable economic growth possible by investment in knowledge capital: Education, R&D increases $ \eta_{tech}$
- India's spending on R & D is only 0.6-0.8 % of GDP, China spends 2.1%, US spends 2.8%, Israel spends 4.3%

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(invest['Year'][36:],invest['R&D'][36:],marker="o")
ax.set_xlabel("Year")
ax.set_ylabel(r"R & D expenditure (as % of GDP)")
ax.set_ylim([0.4,1.0])
ax.set_title("India's R & D expenditure (% of GDP)")
plt.tight_layout()
plt.savefig("Fig5.pdf",dpi=600)
```

# Conclusions

   2. Minimize government investment ($\Delta{K_{public}}$): income tax can be abolished and taxes on fuel can be decreased (Dr. Swamy's advice)
   3. Increase Business Innovation ($\eta_{business}$) by privatisation
   4. Foster Scientific Innovation ($\eta_{tech}$) by huge government spending (6% of GDP)
